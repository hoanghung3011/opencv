```
import cv2 as cv 
import numpy as np
from google.colab.patches import cv2_imshow

# read img
# im = cv.imread('http://otakul.com/uploads/products/5544/Phu-kien-Gundam-Hop-Lon-1.jpg')
im = cv.imread('/content/himekawa.jpg')
cv2_imshow(im)
im = np.float32(im)/255.0

# im = cv.cvtColor(im, cv.COLOR_BGR2GRAY)

# Calculate gradient 
gx = cv.Sobel(im, cv.CV_32F, 1, 0, ksize=1)
gy = cv.Sobel(im, cv.CV_32F, 0, 1, ksize=1)
# https://stackoverflow.com/questions/20979374/opencv-sobel-filter-with-ksize-1
```

```
# Calculate gradient magnitude and direction ( in degrees )
mag, angle = cv.cartToPolar(gx, gy, angleInDegrees=True)
```

```
cv2_imshow(np.float32(mag)*255)
cv2_imshow(np.float32(abs(gx))*255)
cv2_imshow(np.float32(abs(gy))*255)
```
